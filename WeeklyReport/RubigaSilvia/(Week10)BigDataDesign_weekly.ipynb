### Observations
- Most common amino acids in the dataset are Leucine(L), Serine(S), Alanine(A), and Glycine(G). These are known amino acids in proteins and play important roles in protein structure and function
- The last are Cysteine(C), Methionine(M), tryptophan(W), and Histidine(H). They are usually less abundant in proteins but can be important for specific functions.
- Presence of Selenocysteine(U) in the dataset suggests that some of the proteins may be selenoproteins, which contains selenium in the form of selenocysteine instead of cysteine
- Presence of (X,B,Z) and rare amino acids(O,U) suggests that some sequences may be imcomplete or contain errors
